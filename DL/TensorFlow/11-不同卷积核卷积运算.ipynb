```python
import numpy as np

import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
data = plt.imread('./flower.png')[:,:,0]
data.shape
```

```python
plt.imshow(data,cmap = plt.cm.gray)#cmap 
```

```python
input_layer = tf.cast(tf.reshape(data,shape = [1,332,444,1]),dtype=tf.float32)
```

```python
kernel = np.array([[0,0,0],[0,1,0],[0,0,0]])
kernel = tf.cast(tf.reshape(kernel,shape = [3,3,1,1]),dtype=tf.float32)

conv = tf.nn.conv2d(input_layer,kernel,strides=[1,1,1,1],padding = 'SAME')
result = conv.numpy().reshape(332,444)
plt.imshow(result,cmap = 'gray')
```

```python
kernel = np.array([[-1,-1,0],[-1,0,1],[0,1,1]])
kernel = tf.cast(tf.reshape(kernel,shape = [3,3,1,1]),dtype=tf.float32)
conv = tf.nn.conv2d(input_layer,kernel,strides=[1,1,1,1],padding = 'SAME')
result = conv.numpy().reshape(332,444)
plt.imshow(result,cmap = 'gray')
```

```python
kernel = np.array([[1,1,1],[1,-9,1],[1,1,1]])
kernel = tf.cast(tf.reshape(kernel,shape = [3,3,1,1]),dtype=tf.float32)
conv = tf.nn.conv2d(input_layer,kernel,strides=[1,1,1,1],padding = 'SAME')
result = conv.numpy().reshape(332,444)
plt.imshow(result,cmap = 'gray')
```
