```python
import numpy as np

import tensorflow as tf
import matplotlib.pyplot as plt
```

### 激活函数所在的包：neural network

```python
# 线性整流函数：将小于零，变成0，大于0不变
x = np.linspace(-5,10,50)
y = x
y2 = tf.nn.relu(y)
y2 = y2.numpy()
plt.plot(x,y)
plt.plot(x,y2,color = 'red')
```

```python
y3 = tf.nn.sigmoid(y)
y3 = y3.numpy()
plt.plot(x,y3,color = 'r')
plt.plot(x,y)
```

```python
y4 = tf.nn.tanh(y)
y4 = y4.numpy()
plt.plot(x,y4,color=  'red')
plt.plot(x,y)
```
