```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data=[1.0,2.0,3.0]
y_data=[2.0,4.0,6.0]
```

```python
w=torch.Tensor([1.0])
w.requires_grad=True
learn_rate=0.01
```

```python
def forward(x):
    return x*w

def loss(x,y):
    y_=forward(x)
    return (y_-y)**2
```

```python
loss_list=[]
for epoch in range(100):
    for x,y in zip(x_data,y_data):
        l=loss(x,y)
        l.backward()
        print('\t grad:',x,y,w.grad.item())
        w.data-=learn_rate*w.grad.data
        w.grad.data.zero_()
        loss_list.append(l.item())
    print('progress',epoch,l.item())
```
