```python
import matplotlib.pyplot as plt
```

```python
x_data=[1.0,2.0,3.0]
y_data=[2.0,4.0,6.0]
```

```python
w=1.0
learn_rate=0.01
def forward(x):
    return x*w
def loss(x,y):
    y_=forward(x)
    return (y_-y)*(y_-y)
def gradient(x,y):
    return 2*x*(x*w-y)
```

```python
w_list=[]
loss_list=[]
for epoch in range(100):
    for x,y in zip(x_data,y_data):
        grad=gradient(x,y)
        w-=learn_rate*grad
        print('\t grad:',x,y,grad)
        l=loss(x,y)
        w_list.append(w)
        loss_list.append(l)
    print('Epoch',epoch,'w=',w,'loss=',l)
```

```python
plt.plot(w_list,loss_list)
plt.xlabel('w')
plt.ylabel('loss')
```
