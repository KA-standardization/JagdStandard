```python
import numpy as np
import torch
import pandas as pd
```

```python
x_datas=pd.read_csv('./datas/diabetes_data/X.csv',delim_whitespace=True,dtype=np.float32,header=None)
y_datas=pd.read_csv('./datas/diabetes_data/y.csv',delim_whitespace=True,dtype=np.float32,header=None)
```

```python
x_data=torch.from_numpy(np.array(x_datas))
y_d=np.where(np.array(y_datas)>100,1,0)
y_d=y_d.astype(np.float32)
y_data=torch.from_numpy(y_d)
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.linear1=torch.nn.Linear(10,8)
        self.linear2=torch.nn.Linear(8,6)
        self.linear3=torch.nn.Linear(6,4)
        self.linear4=torch.nn.Linear(4,2)
        self.linear5=torch.nn.Linear(2,1)
        self.active=torch.nn.Sigmoid()
        
    def forward(self,x):
        x=self.active(self.linear1(x))
        x=self.active(self.linear2(x))
        x=self.active(self.linear3(x))
        x=self.active(self.linear4(x))
        y_=self.active(self.linear5(x))
        return y_
```

```python
model=Model()
```

```python
criterion=torch.nn.BCELoss(size_average=True)
optimizer=torch.optim.SGD(params=model.parameters(),lr=0.5)
```
