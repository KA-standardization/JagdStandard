```python
# import
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# datas
datas=pd.read_csv('./datas/cancer.csv',sep='\t')
X=datas.iloc[:,2:]
y=datas.iloc[:,1]
datas
```

归一化  
由于特征本身表达方式（度量衡）的原因而导致在绝对数值上的小数据被大数据<font color =red>“吃掉”</font>的情况  
这个时候我们需要做的就是对抽取出来的数据（features vector）进行归一化处理，<font color = red>以保证每个特征被分类器平等对待</font>。

0-1归一化  
$x_{normalization} = \frac{x - min}{max - min}$

```python
# datas
X1=(X-X.min())/(X.max()-X.min())
X_train,X_test,y_train,y_test=train_test_split(X1,y)
```

```python
# knn使用0-1归一化后的数据
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
y_=knn.predict(X_test)
```

```python
# score accuracy
print(knn.score(X_test,y_test))
print(accuracy_score(y_test,y_))
print((y_test==y_).mean())
print(knn.predict_proba(X_test[:5]))
```

Z-score归一化  
$x_{normalization} = \frac{x - \mu}{\sigma}$  
$\mu$ 是平均值 $\sigma$ 是标准差  
方差var，开平方就是标准差  
$\sigma^2 = \frac{\sum(X - \mu)^2}{N}$方差表示数据的离散程度


```python
# datas 每个属性的平均值
X2=(X-X.mean())/X.std()
X_train,X_test,y_train,y_test=train_test_split(X2,y)
X2
```

```python
# knn使用Z-score归一化后的数据
knn.fit(X_train,y_train)
y_=knn.predict(X_test)
```

```python
# score accuracy
print(knn.score(X_test,y_test))
print(accuracy_score(y_test,y_))
print((y_test==y_).mean())
print(knn.predict_proba(X_test[:5]))
```

sigmoid归一化 ----> 0 ~1  
$x_{normalization} = \frac{1}{1 + e^{-x}}$

```python
# datas
sigmoid=lambda x:1/(1+np.exp(-x))
X3=sigmoid(X)
X_train3,X_test3,y_train3,y_test3=train_test_split(X3,y)
X3
```

```python
# knn使用sigmoid归一化后的数据
knn.fit(X_train3,y_train3)
y_3=knn.predict(X_test3)
```

```python
# score accuracy
print((y_test3==y_3).mean())
print(knn.score(X_test3,y_test3))
print(accuracy_score(y_test3,y_3))
print(knn.predict_proba(X_test3[:5]))
```

```python
# 最大值最小值归一化 preprocessing预处理
from sklearn.preprocessing import MinMaxScaler
```

```python
# knn
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
y_=knn.predict(X_test)
```

```python
# accuracy
print(knn.score(X_test,y_test))
print(accuracy_score(y_test,y_))
print((y_test==y_).mean())
```
