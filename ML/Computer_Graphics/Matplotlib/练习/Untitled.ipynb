```python
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
import numpy as np
import scipy.stats as stats
x=np.linspace(-4,4,100)
y=stats.norm.pdf(np.linspace(-4, 4, 100))/np.max(stats.norm.pdf(np.linspace(-3, 3, 100)))
```

```python
#样式 fivethirtyeight
style.use('seaborn')
#画布大小
plt.rcParams['figure.figsize']=(12,6)
#清晰度
plt.rcParams['figure.dpi']=100
#plt.figure(dpi=100)
#设置中文字体以及符号问题
plt.rcParams['font.family']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
plt.fill_between(x,y,alpha=0.5,color='pink')
```

```python
plt.plot(x,y,ls='-.',c='red',lw=5)
```

```python
#注释
plt.plot(x,y,c='red')
plt.plot(x,y**3,c='pink')
plt.text(x=-1.5, y=.7, s="normed", rotation=65, alpha=.75, weight="bold", color="#008fd5")
plt.text(x=-.4, y=.5, s="cdf", rotation=55, alpha=.75, weight="bold", color="#fc4f30")
plt.title('海阔凭鱼越')
plt.legend(['1','2'])
```

```python
#刻度
plt.plot(x,y,c='red')
plt.plot(x,y**3,c='pink')
plt.text(x=-1.5, y=.7, s="normed", rotation=65, alpha=.75, weight="bold", color="#008fd5")
plt.text(x=-.4, y=.5, s="cdf", rotation=55, alpha=.75, weight="bold", color="#fc4f30")
plt.title('海阔凭鱼越')
plt.legend(['1','2'])
plt.tick_params(axis = 'both', which = 'major', labelsize = 18)
plt.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

plt.savefig('雨季.png',dpi=150)
```

```python
#ls='-'是实线,'--'是虚线,':'是点状线,'step'是阶梯线
#marker=默认的是'o'小圆点,'s'小方块,'d'小菱形
plt.plot(x,y,ls=':',marker='s')
```
