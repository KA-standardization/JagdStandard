```python
# 双y轴绘制及合并 多图合并
from matplotlib import rc
# 整体图形设置
rc('mathtext',default='regular')
time=np.arange(10)
temp=np.random.random(10)*30
Swdown=np.random.random(10)*100-10
Rn=np.random.random(10)*100-10
# 画图
fig=plt.figure()
ax=fig.add_subplot(111)
# 画线
ax.plot(time,Swdown,'-',label='Swdown')
ax.plot(time,Rn,'-',label='Rn')
# 双胞胎x轴
ax2=ax.twinx()
ax2.plot(time,temp,'-r',label='temp')
ax.legend(loc=2)
# 网格线
ax.grid()
ax.set_xlabel("Time (h)")
ax.set_ylabel(r"Radiation ($MJ\,m^{-2}\,d^{-1}$)")
# 设置第一个图的纵坐标
ax.set_ylim(-20,100)

# 生成另一个字图形ax2
# 双胞胎x轴，生成另一个图形，ax2，相当于图形ax副本
ax2 = ax.twinx()
ax2.plot(time, temp, '-r', label = 'temp')
ax2.set_ylabel(r"Temperature ($^\circ$C)")
# 设置范围
ax2.set_ylim(0, 35)
# 图例
ax2.legend(loc=0)
plt.savefig('11.png')
```

```python
# 合并图例
rc('mathtext', default='regular')

time = np.arange(10)
temp = np.random.random(10)*30
Swdown = np.random.random(10)*100-10
Rn = np.random.random(10)*100-10

fig = plt.figure()
ax = fig.add_subplot(111)
# 画图接收,接收Lns1列表 对象(就是图形中绘制那条线)
lns1 = ax.plot(time, Swdown, '-', label = 'Swdown')
lns2 = ax.plot(time, Rn, '-', label = 'Rn')
ax2 = ax.twinx()
lns3 = ax2.plot(time, temp, '-r', label = 'temp')

# added these three lines合并列表
lns = lns1+lns2+lns3
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)#向图形中绘制图例,两个图形

ax.grid()
ax.set_xlabel("Time (h)")
ax.set_ylabel(r"Radiation ($MJ\,m^{-2}\,d^{-1}$)")
ax2.set_ylabel(r"Temperature ($^\circ$C)")
ax2.set_ylim(0, 35)
ax.set_ylim(-20,100)
plt.savefig('0.png')
```

```python
# 图形嵌套
%matplotlib inline

# 定义figure
fig = plt.figure()

# 定义数据
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 5, 8, 6]

# figure的百分比, 从figure 10%的位置开始绘制, 宽高是figure的80%
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
# 获得绘制的句柄
ax1 = fig.add_axes([left, bottom, width, height])
# 绘制点(x,y)
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('test')

# 嵌套方法一
# figure的百分比, 从figure 10%的位置开始绘制, 宽高是figure的80%
left, bottom, width, height = 0.2, 0.6, 0.25, 0.25
# 获得绘制的句柄
ax2 = fig.add_axes([left, bottom, width, height])
# 绘制点(x,y)
ax2.plot(x, y, 'r')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('part1')

# 嵌套方法二
plt.axes([bottom, left, width, height])
plt.plot(x, y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('part2')

plt.show()
```

```python
# 模块数据可视化-多图布局，分格显示
%matplotlib inline

plt.figure()
# 通过栅格的形式创建布局方式,(3,3)创建3x3的布局形式，(0,0)绘制的位置，0行0列的位置绘制
# colspan:表示跨几列 rowspan:表示跨几行
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
# 在ax1图中绘制一条坐标(1,1)到坐标(2,2)的线段
# alpha透明度
ax1.plot([1, 2], [1, 2],color='red',alpha=0.1)
# 设置ax1的标题  现在xlim、ylim、xlabel、ylabel等所有属性现在只能通过set_属性名的方法设置
ax1.set_title('ax1_title')  # 设置小图的标题
 
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax3.set_facecolor(color=[1,0,0,0.5])#rgba
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))
# 给对应的图绘制内容，这里只给ax4图绘制，属性通过set_xxx的模式设置
ax4.scatter([1, 2], [2, 2])
ax4.set_xlabel('ax4_x')
ax4.set_ylabel('ax4_y')
plt.show()
```

```python
# 模块数据可视化-多图布局，分格显示subplots
%matplotlib inline

# 这种方法只适合列数相同的布局
# sharex:所有小图共享x轴  sharey:表示所有小图共享y轴  坐标轴以所有小图中范围最大的进行显示
f, ((ax11, ax12), (ax13, ax14)) = plt.subplots(2, 2, sharex=False, sharey=False)
# f就是figure
f.set_figheight(6)
f.set_figwidth(8)
ax11.scatter([1,2], [1,2])
ax12.plot((1,4),(1,4))
x=np.linspace(0,2*np.pi,100)
ax13.plot(x,np.sin(x))
ax14.plot(x,np.cos(x))
# 紧凑显示，边框会比较小，可以注释掉该行查看效果
plt.tight_layout()
plt.show()
```

```python
#gridspec
import matplotlib.gridspec as gridspec
%matplotlib inline 

plt.figure()
# 将整个视图分成3x3布局 网格
gs = gridspec.GridSpec(3, 3)
# gs[0,:]  指定画图的位置 前面指定该图所占的行范围0表示0行，1: 表示从第一行到最后一行
# 第二个参数指定列的范围一个数表示固定列数，x:y表示从x列到y列
ax6 = plt.subplot(gs[0, :])
ax6.plot((0,1),(0,1),color='purple')

# 第一行，从0列开始到2列，不包括2，也就是占0、1两列
ax7 = plt.subplot(gs[1, :2])
ax7.set_facecolor(color='green')
# 从第一行到最后，占1、2两行，后面的2表示只占用第二列，也就是最后的一列
ax8 = plt.subplot(gs[1:, 2])
# 倒数第一行，只占第0列这一列
ax9 = plt.subplot(gs[-1, 0])
# 倒数第一行，只占倒数第二列，由于总共三列，所以倒数第二列就是序号1的列
ax10 = plt.subplot(gs[-1, -2])
plt.show()
```

```python
# 热力图
# sphinx_gallery_thumbnail_number = 2

# 数据，存在关系
# 热力图 表示两种属性之间的关系的一种图形
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
# 农民，生产不同的蔬菜，常量，每个人各不相同，如果数值是0，表示，不种这种蔬菜
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

# 收获
harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])

# 此时显示的图片就是这个对象，设置对象的宽度和高度，图片放大
fig, ax = plt.subplots()
fig.set_figheight(6)
fig.set_figwidth(6)
im = ax.imshow(harvest,cmap =plt.cm.YlGn )#imshow显示图片

# We want to show all ticks... 刻度
ax.set_xticks(np.arange(len(farmers)))
ax.set_yticks(np.arange(len(vegetables)))
# ... and label them with the respective list entries
# 方式一
ax.set_xticklabels(farmers,rotation = 60,rotation_mode = 'anchor',ha = 'right',fontsize = 15)
ax.set_yticklabels(vegetables,color = 'red')

# 方式二，横坐标的刻度进行旋转。
# setp -----> properties设置属性
# plt.setp(ax.get_xticklabels(), rotation=45, ha="right",rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
#         图形中绘制文本
# ha horizontal align 水平对齐方式
# va vertical align   竖直对齐方式
# 坐标系和平时的坐标系，正好相反的！
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="k")

# 设置标题
ax.set_title("Harvest of local farmers (in tons/year)")
# tight紧凑
fig.tight_layout()
plt.colorbar(im,shrink = 0.6)# color bar，颜色条。
```

```python
# 轮廓图
from matplotlib import cm

# 构造数据
delta = 0.5
extent = (-3, 4, -4, 3)
x = np.arange(-3.0, 4.001, delta)
y = np.arange(-4.0, 3.001, delta)
X, Y = np.meshgrid(x, y)# 网格交叉，大家下课自己测试
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

# Boost the upper limit to avoid truncation errors.
levels = np.arange(-2.0, 1.601, 0.02)
norm = cm.colors.Normalize(vmax=abs(Z).max(), vmin=-abs(Z).max())
cmap = cm.PRGn#颜色

# 添加子视图
fig, axs = plt.subplots(nrows=1, ncols=1)

# 轮廓面
cset1 = axs.contourf(X, Y, Z, levels, norm=norm,
                     )
# 轮廓线----->等高线
cset2 = axs.contour(X, Y, Z, cset1.levels)

for c in cset2.collections:
    c.set_linestyle('-.')
    

cset3 = axs.contour(X, Y, Z, (0,), colors='g', linewidths=5)

axs.set_title('Filled contours')
fig.colorbar(cset1, ax=axs)
fig.tight_layout()
plt.show()
```

```python
# 3D
from mpl_toolkits.mplot3d.axes3d import Axes3D


fig = plt.figure(figsize=(8,8))

# 构建三维图形
ax = fig.add_subplot(1, 1, 1, projection='3d')

X = np.arange(-5, 5, 0.25) #横坐标
Y = np.arange(-5, 5, 0.25) #纵坐标
X, Y = np.meshgrid(X, Y)# 面交叉
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# 绘制面
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                       linewidth=0, antialiased=False)
ax.set_zlim(-1, 1)

plt.show()
```
