```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans,DBSCAN
from sklearn import datasets
```

```python
# datas
X,y=datasets.make_circles(n_samples=1000,
                          shuffle=True,
                          factor=0.5,#距离大小
                          noise=0.05)#高斯噪声
X1,y1=datasets.make_blobs(n_samples=100,
                          n_features=2,
                          centers=[(1.5,1.5)],#中心点的坐标值
                          cluster_std=0.2)
plt.scatter(X[:,0],X[:,1],c=y)
plt.scatter(X1[:,0],X1[:,1],c='g')
```

```python
X3=np.concatenate([X,X1])
y3=np.concatenate([y,y1+2])
plt.scatter(X3[:,0],X3[:,1],c=y3)
```

```python
k=KMeans(n_clusters=3).fit(X3)
plt.scatter(X3[:,0],X3[:,1],c=k.labels_)
```

```python
scan=DBSCAN(eps=0.2,min_samples=5).fit(X3)
plt.scatter(X3[:,0],X3[:,1],c=scan.labels_)
```
