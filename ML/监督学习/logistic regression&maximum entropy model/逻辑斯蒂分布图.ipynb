```python
# import
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(-8,8)
```

```python
fun=lambda x:1/(1+np.exp(-x))
y=np.array(fun(X))
```

```python
X_new=np.linspace(X.min(),X.max(),300)
_=plt.plot(X,y)
```

```python
fun2=lambda x:0.5*(np.exp(-x)/(1+np.exp(-x))**2)
y2=fun2(X)
```

```python
_=plt.plot(X,y2)
```
