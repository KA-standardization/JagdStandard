```python
# import
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree,datasets
from sklearn.ensemble import AdaBoostClassifier
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# datas
X,y=datasets.load_iris(True)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1024)
```

```python
# 箱式图
_=plt.boxplot(X)
```

```python
# AdaBoostClassifier
ada=AdaBoostClassifier(n_estimators=3,
                       algorithm='SAMME',
                       learning_rate=0.5)
ada.fit(X_train,y_train)
y_=ada.predict(X_test)
ada.score(X_test,y_test)
```

```python
display(y_)
display(ada.predict_proba(X_test))
```

```python
# 树
_=tree.plot_tree(ada[0],filled=True)
```

softmax  转化概率
$p=\frac{e^X}{e^X.sum()}$

```python
X=np.random.randint(0,10,size=5)
print(X/X.sum())
X
```

```python
# softmax 非线性 把小的变得更小 大的变得更大
p=(np.e**X)/((np.e**X).sum())
print(p.sum())
p
```
